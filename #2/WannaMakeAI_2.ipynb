```python
import numpy as np
import matplotlib.pyplot as plt
import decimal

class ActivationFunctions:
  def __init__(self):
    self.x = [(i) for i in range(-100,100)]

  # Binary Step Function
  def binary_step(self, threshold=5):
    result = []
    for i in (self.x):
      if i>threshold:
        result.append(1)
      else:
        result.append(0)
    return result

  # Linear Function
  def linear(self, m=12, c=5):
    result = []
    for i in (self.x):
      result.append(m*(i)+c)
    return result

  # Sigmoid function
  def sigmoid(self):
    result = []
    for i in self.x:
        result.append(1/(1 + np.exp(-i)))
    return result

  # Tanh Function
  def tanh(self):
    result = []
    for i in self.x:
        result.append( (2/(1 + np.exp(-2*i))) - 1 )
    return result

  # ReLU Function
  def relu(self):
    result = []
    for i in self.x:
        if(i<0):
          i=0
        result.append(i)
    return result
  
  # Leaky-ReLU Function
  def leaky_relu(self):
    result = []
    for i in self.x:
        if(i<0):
          i=0.01*i
        result.append(i)
    return result

  # Maxout Function
  def maxout(self):
    return "\nThis is a complex implementation and not shown here \
            \nBut you can check it out here: \
            \nhttps://github.com/philipperemy/tensorflow-maxout/blob/master/maxout.py"

  # Parameterised ReLU Function
  def param_relu(self, a=0.1):
    result = []
    for i in self.x:
        if i<0:
            i = a*i
        result.append(i)
    return result

  # Exponential Linear Unit
  def elu(self, a=0.5):
    result = []
    for i in self.x:
        if(i<0):
          i=a*(np.exp(i)-1)
        result.append(i)
    return result  

  # Softmax function
  def softmax(self):
    phi = np.exp(self.x)
    phi_T = phi/phi.sum()
    return phi_T

  # Swish function
  def swish(self):
    result = []
    for i in self.x:
        result.append(i/(1+np.exp(-i)))
    return result


  # Plot the Graph
  def plot_func(self, y, tlt):
    plt.plot(self.x, y)
    plt.title(tlt)
    plt.show()

  # Plot All
  def plot_all(self):
    figure, axis = plt.subplots(2, 5)
    axis[0,0].plot(self.x, self.binary_step())
    axis[0,0].set_title("Binary step Function")
    axis[0,1].plot(self.x, self.linear())
    axis[0,1].set_title("Linear Function")
    axis[0,2].plot(self.x, self.sigmoid())
    axis[0,2].set_title("Sigmoid Function")
    axis[0,3].plot(self.x, self.tanh())
    axis[0,3].set_title("Tanh Function")
    axis[0,4].plot(self.x, self.relu())
    axis[0,4].set_title("ReLU Function")
    axis[1,0].plot(self.x, self.leaky_relu())
    axis[1,0].set_title("Leaky-ReLU Function")
    axis[1,1].plot(self.x, self.param_relu())
    axis[1,1].set_title("Parameterised ReLU Function")
    axis[1,2].plot(self.x, self.elu())
    axis[1,2].set_title("ELU Function")
    axis[1,3].plot(self.x, self.softmax())
    axis[1,3].set_title("Softmax Function")
    axis[1,4].plot(self.x, self.swish())
    axis[1,4].set_title("Swish Function")
    # set the spacing between subplots
    plt.subplots_adjust(left=1, bottom=1,right=3,top=2,wspace=0.5,hspace=0.5)
    plt.show()
  
  # # Compare Graphs
  # def compare_graphs(self):
  #   plt.plot(self.x, self.binary_step(), label='Binary Step')
  #   plt.plot(self.x, self.linear(), label='Linear')
  #   plt.plot(self.x, self.sigmoid(), label='Sigmoid')
  #   plt.plot(self.x, self.tanh(), label='Tanh')
  #   plt.plot(self.x, self.relu(), label='ReLU')
  #   plt.plot(self.x, self.leaky_relu(), label='Leaky ReLU')
  #   plt.plot(self.x, self.param_relu(), label='Param ReLU')
  #   plt.plot(self.x, self.elu(), label='ELU')
  #   plt.plot(self.x, self.softmax(), label='Softmax')
  #   plt.plot(self.x, self.swish(), label='Swish')
  #   plt.legend()
  #   plt.show()

if __name__ == '__main__':  

  func = ActivationFunctions()

  while True:
    try:
      ip =int(input("\
      \nChoose an option from below:\
      \n1)Binary Step Function\
      \n2)Linear Function\
      \n3)Sigmoid Function\
      \n4)Tanh Function\
      \n5)ReLU Function\
      \n6)Leaky-ReLU Function\
      \n7)Maxout Function\
      \n8)Parameterised Function\
      \n9)Exponential Linear Unit Function\
      \n10)Softmax Function\
      \n11)Swish Function\
      \n12)Plot All Graphs Separately\
      \n13)Exit\n"))
    except:
      print("\n/**********/\nPlease Enter only integer values\n/**********/\n")

    
    
    if ip==13:
      print("Exiting the Application...")
      break

    if ip==12:
      func.plot_all()

    if ip==1:
      threshold = int(input("\nPlease enter a threshold value:\n"))
      y = func.binary_step(threshold)
      tlt="Binary Step Function"
      func.plot_func(y, tlt)
    if ip==2:
      m = int(input("\nPlease enter the slope value:\n"))
      c = int(input("\nPlease enter the constant value:\n"))
      y = func.linear(m, c)
      tlt="Linear Function"
      func.plot_func(y, tlt)
    if ip==3:
      y = func.sigmoid()
      tlt="Sigmoid Function"
      func.plot_func(y,tlt)
    if ip==4:
      y = func.tanh()
      tlt="tanh Function"
      func.plot_func(y,tlt)
    if ip==5:
      y = func.relu()
      tlt="ReLU Function"
      func.plot_func(y,tlt)
    if ip==6:
      y = func.leaky_relu()
      tlt="Leaky ReLU Function"
      func.plot_func(y,tlt)
    if ip==7:
      y = func.maxout()
      tlt="Maxout Function"
      print(y)
    if ip==8:
      print("\nParameter value used is: 0.1\n")
      y = func.param_relu()
      tlt="Parameterised ReLU Function"
      func.plot_func(y,tlt)
    if ip==9:
      print("\nParameter value used is: 0.5\n")
      y = func.elu()
      tlt="Exponential Linear Unit Function"
      func.plot_func(y,tlt)
    if ip==10:
      y = func.softmax()
      tlt="Softmax Function"
      func.plot_func(y, tlt)
    if ip==11:
      y = func.swish()
      tlt="Swish Function"
      func.plot_func(y,tlt)
```
